```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('housing.csv')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['median_house_value'],axis=1)
```

```python
X.head()
```

```python
Y=data['median_house_value']
```

```python
Y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms'])
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
train_data.head()
```

```python
train_data.hist(figsize=(15,10))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
ocean_prox=pd.get_dummies(train_data.ocean_proximity)
```

```python
train_data=train_data.join(ocean_prox).drop(columns=['ocean_proximity'])
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=train_data,hue='median_house_value',palette='coolwarm')
```

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,y_train=train_data.drop(columns=['median_house_value']),train_data['median_house_value']
```

```python
linear=LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
test_data=x_test.join(y_test)

test_data['total_rooms']=np.log(test_data['total_rooms'])
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)

ocean_prox=pd.get_dummies(test_data.ocean_proximity)
test_data=test_data.join(ocean_prox).drop(columns=['ocean_proximity'])

test_data['bedroom_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']
x_test,y_test=test_data.drop(columns=['median_house_value']),test_data['median_house_value']
```

```python
linear.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest=RandomForestRegressor()
```

```python
forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_s=scaler.fit_transform(x_train)
```

```python
forest.fit(x_train_s,y_train,n_estimators=200)
```

```python
x_test_s=scaler.fit_transform(x_test)
```

```python
forest.score(x_test_s,y_test)
```
